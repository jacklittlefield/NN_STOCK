```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as datetime
```

```python
df = pd.read_csv('AAPL (2).csv')
```

```python
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
#df.drop('date', axis = 1, inplace=True)
```

```python
#df.head()
```

```python
df.set_index(df['Date'], inplace=True)
```

```python
plt.plot(df['Adj Close']);
```

```python
df['std'] = df['Adj Close'].rolling(20).std()
```

```python
df.tail()
```

```python
df['upper'] = df['Adj Close'] + (df['std']*2)
df['lower'] = df['Adj Close'] - (df['std']*2)
```

```python
plt.figure(figsize=(20,10))
plt.plot(df[['Adj Close', 'upper', 'lower']]);
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['std']);
```

```python
plt.figure(figsize=(20,10))
plt.plot(df['Volume']);
```

```python
df.tail()
```

```python
df.drop(['Open', 'High', 'Low', 'Close', 'Date'], axis = 1, inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
#df.drop('Date', axis =1, inplace=True)
```

```python
df.head()
```

```python
df['Adj Close'].min()
```

```python
rates = pd.read_csv('TB3MS.csv')
```

```python
rates.head()
```

```python
rates['Date'] = pd.to_datetime(rates['DATE'])
```

```python
rates.set_index('Date', inplace=True)
```

```python
rates.drop('DATE', axis=1, inplace=True)
```

```python
names = {'TB3MS':'Rates'}
```

```python
rates.rename(mapper=names, axis = 1, inplace=True)
```

```python
rates
```

```python
plt.figure(figsize=(20,10))
plt.plot(rates);
```

```python
comb = pd.concat((df, rates), axis=1)
```

```python
comb.head()
```

```python
comb['Rates'] = comb['Rates'].fillna(method='ffill')
```

```python
comb.tail()
```

```python
comb.dropna(inplace=True)
```

```python
comb.head()
```

```python
plt.figure(figsize=(20,10))
plt.plot(comb[['Rates', 'std']]);
```

```python
comb.to_csv('aapl_rates.csv', index=True, header=True, date_format='%Y-%m-%d')
```
