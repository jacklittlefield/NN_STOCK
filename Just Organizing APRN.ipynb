```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('APRN.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df1 = pd.read_csv('aprn_rates_nonadj.csv')
df1['Date'] = pd.to_datetime(df1['Date'])
df1.set_index('Date', inplace=True)
```

```python
comb = pd.concat((df,df1), axis = 1)
```
