```python
def pe_ratio(str):
    ticker = str.lower()
    try:
        url = 'https://finviz.com/quote.ashx?t='+ticker
        res = requests.get(url)
        soup = bs(res.content, 'lxml')
        pe =  soup.find(text = r'P/E')
        pe_ = pe.find_next(class_='snapshot-td2').text
        print('The Price/Earnings Value for '+str+' is', pe_)
        return pe_
    except exception as e:
        print(e)
```
