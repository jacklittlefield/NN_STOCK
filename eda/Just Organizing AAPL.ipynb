```python
import pandas as pd
import datetime
```

```python
df = pd.read_csv('APRN.csv')
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
df.head()
```

```python
df1 = pd.read_csv('aapl_rates.csv')
df1['Date'] = pd.to_datetime(df1['Date'])
df1.set_index('Date', inplace=True)
```

```python
comb = pd.concat((df,df1), axis = 1)
```

```python
comb.columns
```

```python
comb = comb.drop(comb[['Open', 'High', 'Low', 'Adj Close', 'Volume', 'Adj Close',
       'Volume', 'std', 'upper', 'lower']], axis = 1)
```

```python
comb.dropna(inplace=True)
```

```python
comb.head()
```

```python
pd.DataFrame.to_csv(comb, 'APRN_COMB.csv')
```
